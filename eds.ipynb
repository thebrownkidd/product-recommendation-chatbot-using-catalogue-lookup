## Categories
few ideas here:
1. make this a semantic feature, vectore encode this, then match with the title. make a siimilarity score between categories and titlee + discriptions
2. Cosine similarity search based on this column alone for higher than primar feature status.

```python
cats = {}
for i in range(len(df)):
    for cat in ut.MakeList(df["categories"].tolist()[i]):
        cats[cat.strip()] = cats.get(cat.strip(), 0) + 1
# Sort categories by frequency
print(f"Total Categories : {len(cats)}")
print("\n\n")
sorted_cats = sorted(cats.items(), key=lambda x: x[1], reverse=True)
for cat, freq in sorted_cats:
    print(f"{cat}: {freq}")
```
